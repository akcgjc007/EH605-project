```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./data_GUJARAT_23.125_72.375.txt',
                 delimiter=' ', header=None)

for year in range(2000, 2005):
        x = df.loc[df[0] == year].index
        y = df.loc[df[0] == year][18]*(1/(df.loc[df[0] == year][3]+273.15) - 1/5800)
        plt.plot(x, y)
plt.xlabel("Year")
plt.ylabel("Entropy")
plt.legend([2000+i for i in range(5)])
plt.show()
```
